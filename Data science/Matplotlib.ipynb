# Matplotlib

Biblioteka `matplotlib` jest podstawową biblioteką do tworzenia wizualizacji w Pythonie. Jego największymi zaletami są uniwersalność i powszechność - w internecie znajdziemy mnóstwo informacji na temat tego jak używać tego pakietu i tworzyć w nim wizualizacje.

Jest uważana za "dziadka" tego typu narzędzi w Pythonie. Została stworzona przez Johna Huntera w celu odtworzenia możliwości tworzenia wykresów znanych z MatLaba (innego języka programowania) w Pythonie. Jeśli więc znasz Matlaba, korzystanie z Matplotlib będzie dla ciebie naturalne.

Inne biblioteki to: Seaborn oraz Plotly.

To doskonała biblioteka do tworzenia grafiki 2D i 3D do generowania wykresów naukowych.

Oto kilka głównych zalet Matplotlib:

- Ogólnie łatwa inicjacja w przypadku prostych wykresów
- Wsparcie dla niestandardowych etykiet i tekstów
- Doskonała kontrola nad każdym elementem na wykresie
- Wysokiej jakości generowanie wyjścia w wielu formatach
- Ogólna możliwość dużego dostosowania

Matplotlib pozwala tworzyć replikowalne wykresy programowo. Nauczmy się, jak z niej korzystać! Oficjalna strona internetowa Matplotlib: https://matplotlib.org/

```python
fig, ax = plt.subplots(2)

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```

Jeżeli chcemy kilka serii danych (wykresów) na raz możemy używać `plt.plot` wiele razy:

```python
plt.plot(x, np.sin(x))
plt.plot(x, 2*np.sin(x))
```

Aby zarządzać kolorami możemy użyć parametru `color`, oto kilka sposobów używania go:

```python
plt.plot(x, np.sin(x - 0), color='purple')      # nazwa koloru
plt.plot(x, np.sin(x - 1), color='r')           # kod koloru (rgbcmykw)
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # hex RRGGBB
plt.plot(x, np.sin(x - 4), color=(0.65, 0.6, 0.5)) # trójka reprezentująca wartości RGB (red, green, blue)
```

Analogicznie możemy definiować styl linii za pomocą `linestyle`

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# Lub korzystając ze skrótów:
plt.plot(x, x + 4, linestyle='-')  
plt.plot(x, x + 5, linestyle='--') 
plt.plot(x, x + 6, linestyle='-.') 
plt.plot(x, x + 7, linestyle=':');
```

Możemy też łączyć ze sobą te komendy:

```python
plt.plot(x, x + 0, 'g-')
plt.plot(x, x + 1, '--c')
plt.plot(x, x + 2, '-.k')
plt.plot(x, x + 3, 'r:');
```

### Ustawianie zakresu osi

```python
plt.plot(x, np.sin(x), "-b")

plt.xlim(-2, 12)  # zakres osi OX
plt.ylim(-1.5, 1.5);  # zakres osi OY
```

### Dodawanie oznaczeń do wykresów

```python
plt.plot(x, np.sin(x))
plt.title("A Sin Curve")
plt.xlabel("x")  
plt.ylabel("sin(x)");
```

Kiedy pokazujemy więcej niż jedną serie danych możemy skorzystać z `plt.legend` aby dodać legendę do wykresu:

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('tight')

plt.legend();
```

```python
import matplotlib.pyplot as plt
import pandas as pd

x = np.linspace(0, 10, 100)
y = (x + 11) / 3

df = pd.DataFrame(data={"x": x,
                        "y": y})

plt.plot(df.loc[:, "x"], df.loc[:, "y"], '--c', label="my plot")
plt.xlabel("x")
plt.ylabel("y")
plt.title("My plot")
plt.xlim(-10, 19)
plt.ylim(0, 10)

plt.legend(loc='best')
```

Jeżeli korzystamy z obiektowego interfejsu nazwy metod mogą być trochę inne:

- `plt.xlabel()` → `ax.set_xlabel()`
- `plt.ylabel()` → `ax.set_ylabel()`
- `plt.xlim()` → `ax.set_xlim()`
- `plt.ylim()` → `ax.set_ylim()`
- `plt.title()` → `ax.set_title()`


Aby ułatwić sobie pracę możemy skorzystać z `ax.set` i ustawić wszystko za jednym razem:

```python
fig = plt.figure()
ax = plt.axes()

ax.plot(df.loc[:, 'x'], df.loc[:, 'y'])

ax.set(xlim=(0, 10), ylim=(3, 7),
       xlabel='x', ylabel='sin(x)',
       title='A Simple Plot');
```

### Wykresy kropkowe (scatter plots)

Wykresy kropkowe tworzymy dokładnie tak jak liniowe, zmieniamy jedynie symbol który jest używany do rysowania:

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, "ko");
```

Co więcej symbole można ze sobą łączyć:

```python
plt.plot(x, y, 'r*-');
# plt.plot(x, y, linestyle="solid", color="red", marker="*")
```

Funkcja `plot` posiada wiele argumentów służacych do modyfikowania stylu linii i punktów:

```python
plt.figure(figsize=(10,10))
plt.plot(x, y, linestyle='solid',
         marker='s', color='gray',
         markersize=6, linewidth=4,
         markerfacecolor='blue',
         markeredgecolor='red',
         markeredgewidth=2
        )
plt.axis('tight');
```

Jeżeli chcemy tworzyć bardziej skomplikowane wykresy kropkowe, możemy skorzystać z `plt.scatter` które daje nam możliwość ustalania stylu dla każdej obserwacji:

```python
plt.scatter?
```

```python
plt.scatter(x, y, c='yellow')
```

### Tworzenie histogramów

Histogramy są świetnym narzędziem na początkowym etapie analizy zbioru danych, aby lepiej zrozumieć ich rozkład.

```python
data = np.random.randn(1000)
plt.hist(data);
```

```python
data[:10]
```

```python
plt.hist(data, bins=30, alpha=0.5, histtype='bar', color='steelblue', edgecolor='red');
```

```python
plt.hist?
```

```python
x1 = np.random.normal(0, 0.5, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

kwargs = dict(histtype='bar', alpha=0.3, bins=40)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs);
```

### Tworzenie boxplotów

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

plt.boxplot(data, vert=True, patch_artist=True);   
```
